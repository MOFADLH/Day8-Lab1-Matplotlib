```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Q1: Create two charts using list and dataframe

```python
plt.plot([1,2,0,4,5])
plt.plot([9,8,2,3,6])
```

## Q2: Create pie chart with equal explode values.

```python
y = np.array([35, 25, 25, 15])
mylabels = ['Apples', 'Tomatoes','Mangoes','Oranges']
myexplode = [0.2, 0.2, 0.2,  0.2]

plt.pie(y, labels = mylabels, explode = myexplode)
plt.show() 
```

## Q3: Create a scatter chart with circle marker, alpha = 0.4, and size = 50.

```python
x = np.random.randn(600)
y = np.random.randn(600)
plt.scatter(x, y,alpha =0.4, marker = 'o' ,s = 50)
```

## Q4: Add the following to your chart:
- title placed in the left and has a red color
- x and y label with font size = 16

```python
plt.pie(np.arange(1,5),labels=['Apples', 'Tomatoes','Mangoes','Oranges']);
plt.title('fruits', fontsize = 16, loc='left', color='r');
```

## Q5: Create three line charts with Legends at left and colors

```python
plt.plot([1,2,0,5,6])
plt.plot([1,3,3,8,6])
plt.plot([1,4,7,5,9])
plt.legend(['1','2','3'], shadow = True, loc =1)
```
